Write a program to calculate the electricity bill (accept number of unit from user) according to the following criteria:

  </tr>

  </tr>
  <tr>
    <td>Next 100 units</td>
    <td colspan = "2">Rs 5 per unit</td>

    <td colspan = "2">Rs 10 per unit</td>
    
  </tr>
</table> <br/>
    (For example if input unit is 350 than total bill amount is Rs2000)




```python
unit = int(input("Enter the units: "))
if unit <= 100:
    print("No charge")
elif unit > 100 and unit <= 200:
    charge = (unit-100) * 5
    print("your cost is Rs ",charge)
elif unit > 200:
    charge = ((unit-200) * 10) + (200-100) * 5 #we are deducting 100 from 200, (200-100), because first 100 units is free of charge
    print("your cost is Rs ",charge)

```
