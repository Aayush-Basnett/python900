# Find_all and filter
The find_all method and the filter method are similar in that they both allow you to search for elements in a BeautifulSoup object that match a certain criteria. However, they have some differences.

find_all is a BeautifulSoup method that returns a list of all the matching tags and their contents. It allows you to specify a tag name, a list of tag names, a class, an id, or any combination of these as arguments. For example, soup.find_all('p') will return a list of all the `<p>` tags in the soup.

filter is a Python built-in function that takes a function and an iterable as arguments, and returns an iterator containing all the elements of the iterable for which the function returns True. In BeautifulSoup, you can use the filter function to filter the contents of a tag or a list of tags based on some criterion. For example, filter(lambda x: 'spam' in x, soup) will return an iterator containing all the elements of the soup that contain the string 'spam'.

So while they are similar in some ways, they have different use cases. find_all is more commonly used in BeautifulSoup to find elements that match a certain criteria, while filter is more general-purpose and can be used in a wider variety of situations.

```python
# now we loop through all the rows starting from [8] to get all rows
all_data = []
for row in range(8, len(table_row)): #this loop is to get all the rows
    #we have now entered an individual row
    cd = [] 
    country_data = table_row[row].find_all("td") #this finds all the table data inside of the specific row
 # This loop appends the text content of each table cell to the 'cd' list
    # It also removes some of the unnecessary tags and attributes    
    for item in country_data:
        print(item.text)
        cd.append(item.text)
    all_data.append(cd)
    
```

```python
# To make the data organized we use pandas
import pandas as pd
pd.DataFrame(all_data)
```
