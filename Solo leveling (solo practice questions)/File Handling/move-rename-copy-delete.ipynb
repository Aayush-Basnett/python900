## Rename


```python
for file in os.listdir():
    if file == ".DS_Store": #.DS_Store is a hidden file in macOs which makes the folder it is contained in accesable by the finder app
        continue
    name, ext = os.path.splitext(file)
    splitted = name.split(" ")
    # Lets get rid of unnecessary spaces
    splitted = [string.strip() for string in splitted]
    new_name = f"{splitted[1]}-{splitted[0]}{ext}"  
    # Changed from name number.extention to number-name.extention
    os.rename(file, new_name)
    
```

## Move


```python
#for this first we gotta import shutil module
import shutil
print(os.getcwd())
```

```python
#Lets move the files we just renamed into a new folder
if not os.path.exists("data"):
    os.mkdir("data")
    #This code states that if a folder named data does't exist then a new folder will be created
    print("New folder created")
else:
    print("folder already exists")


```

```python
for file in os.listdir():
    if file == ".DS_Store":
        continue
    shutil.move(file, "data") #move the file to folder we just created
```
