# Git for dummies


Git is basically like a memorycard for all your code which can also be accessed by other people if you wish to do so. 


If you are playing a video game, and then after making some progress let's say your character dies, then if you have saved your progress earlier then you won't lose all your progress, you won't need to start the game from scratch. This same concept applies for git. 

Say that after writing a code you make an error and you want to travel back in time and rewrite a new code, git allows you to do so. 

```git
git init
```
This is like inserting a memory card into your computer


```git
git add {file name or . to add all}
```
add is used to select the files that we want to save, it could be a specific file or everything {represented by `.`}

```git
git commit -m "your message"
```
This commands adds the files that we previously selected using `git add` to the memory card{analogy}

# Branches
Lets say that you are play a video game, your little brother also wants to play it but he sucks at that game. So, you don't want him to ruin your progress but he wants to play with the characters and tools that you have unlocked. Then you can make a new branch for him to play the game and even if he ruins everything then the changes won't be reflected in your code. 

This is the same concept of branches in git.
The following is the code used to make a branch

```git
git checkout -b {branch name}
```
let's say the branch name is called `new-branch`. Then, any changes you made in the new-branch branch will not be reflected in master branch

Unless the new branch is merged to main/master, the main code will be intact. 

To push the new branch to github, use the following code

```git
git push origin {branch name}
```
if branch name is called `new-branch`, a new branch will be added to github, duh.



To switch branches, use following code

```git
git checkout {branch name}
```

# So all in all
```git
git init = initialize project to use git
git add .= add all changes to be saved
git add *filename* = add single file to be saved
git commit -m "message" = savechanges with message
git push origin master = push changes to github master
git push origin new-branch = push changes to github new-branch
git pull origin master = pull changes from github master
git checkout -b new-branch = create a new branch
git status = check status of changes
git log = see all previous saved changes
git checkout *commit hash* = travel back to old commit
```
