# Assignment -2
use python intrepretor to calculate the following
* how many seconds in 42 min and 42 sec?
* how many miles in 10km ?(1mile = 1.609km)
* if you run a 10km race in 42 min 42 sec, whats your average pace? (speed : miles in minutes and miles in seconds)

```python
OneMin = 60
fourtyTwoMin = OneMin * 42
TotalSeconds = fourtyTwoMin + 42
print(f"there are {TotalSeconds} seconds in 42 minutes and 42 seconds")
```

```python
OneKM = 1/1.609
tenKM = OneKM * 10
print(f"there are {tenKM} miles in 10km")
```

```python
averageSpeed_miles_in_minutes = tenKM/(TotalSeconds/60)
averageSpeed_miles_in_seconds = tenKM/TotalSeconds

print(f"hence, average speed in miles per minute is {averageSpeed_miles_in_minutes}mil/min and in miles per second is {averageSpeed_miles_in_seconds}mil/sec")
```

# Assignment 3
Volume of sphere with radius r is $\frac{4}{3}Ï€r^3$. What is the volume of sphere with radius 5?

```python
#solution
vol = 4/3*3.14*int(input("declare value of r: "))**3
round_off_value = round(vol, 2)
print(vol, "but round off is ", round_off_value)
```

# Alternate Solutions assignment 1



```python
num1, num2= 2.5, 0.36
print(f"before num1={num1} and num2={num2}")
num1, num2= 0.36, 2.5
print(f"after num1={num1} and num2={num2}")

```

```python
num1, num2= 2.5, 0.36
print(f"before num1={num1} and num2={num2}")
tmp = num1
num1= num2
num2 = tmp
print(f"after num1={num1} and num2={num2}")
```
